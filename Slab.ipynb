```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e1p = 1
e1max = e1p *np.conjugate(e1p)
rmr = 1.5
rmi = 0.05
m=complex(rmr,rmi)
```

```python
xmin=0.0
xmax=10.0
npoints=500
x=np.linspace(xmin,xmax,npoints)
```

```python
ei4mx = np.exp(1j*4*m*x)
denom = (m*m+1)*(ei4mx - 1.0)-2.0*m*(ei4mx +1)
e1m = e1p*(m*m-1)*(1-ei4mx)/denom
e3p = e1p*(-4)*m*np.exp(1j*2*x*(m-1))/denom
```

```python
eerefl = e1m*np.conjugate(e1m)/e1max # Reflected intensity
eetrns = e3p*np.conjugate(e3p)/e1max # Transmitted intensity
```

```python
#Intensity Plot
plt.plot(x,eerefl)
plt.plot(x,eetrns)
plt.xlabel("Size parameter", fontsize=14)
plt.ylabel("Intensity", fontsize=14)
plt.show()
```
